```python
import numpy as np
import pandas as pd
import openpyxl
```

```python
df=pd.read_excel("responses.xlsx")
df.head()
```

```python
# Define the function to categorize locations
def categorize_location(location):
    if location == "Isb/Pindi,  Lahore,  Karachi":
        return "Urban Professional Market"
    elif location == "Faisalabad, Hyderabad, Multan,Quetta, Peshawar, Gilgit , etc":
        return "Emerging Urban Market"
    elif pd.isna(location) or location == "":
        return "Not Specified"
    else:
        return "Rural Market"  # For any other unexpected values

# Apply the function to create the new column
df["respondent_location"] = df["Where do you live"].apply(categorize_location)

# Display the resulting DataFrame
df[["Where do you live", "respondent_location"]]
```

```python
# Define the conditions and corresponding values
conditions = [
    (df['If your hair care routine were a story, what would it be?   '] == "Quick morning action movie"),
    (df['If your hair care routine were a story, what would it be?   '] == "A relaxing self-care spa experience"),
    (df['If your hair care routine were a story, what would it be?   '] == "A scientific experiment"),
    (df['If your hair care routine were a story, what would it be?   '] == "A nature-inspired wellness journey"),
    (df['If your hair care routine were a story, what would it be?   '] == "A budget-conscious survival tale")
]

choices = [
    "Time-Pressed",
    "Relaxation-Seeking",
    "Exploration-Driven",
    "Nature inspired",
    "Cost-Aware"
]

# Create the new column
df['hair_care_routine_story'] = np.select(conditions, choices, default='Other')

# Display the resulting DataFrame
df[['If your hair care routine were a story, what would it be?   ', 'hair_care_routine_story']]
```

```python
#remove quotes from the beginning and end of the string
df['Your ideal hair care product speaks to you. What does it say? '] = df['Your ideal hair care product speaks to you. What does it say? '].str.strip('"')
```

```python
# Define the conditions and corresponding values
conditions_message = [
    (df['Your ideal hair care product speaks to you. What does it say? '] == "I'll make you look Instagram-perfect"),
    (df['Your ideal hair care product speaks to you. What does it say? '] == "I'll keep you healthy naturally"),
    (df['Your ideal hair care product speaks to you. What does it say? '] == "I understand your unique hair story"),
    (df['Your ideal hair care product speaks to you. What does it say? '] == "I'm gentle and kind to you")
]

choices_message = [
    "Image-Driven",
    "Naturally Healthy",
    "Personalization-Focused",
    "Gentle-Caring"
]

# Apply np.select to create the new column based on the conditions
df['ideal_hair_care_product_message'] = np.select(conditions_message, choices_message, default='Other')

df[['Your ideal hair care product speaks to you. What does it say? ', 'ideal_hair_care_product_message']]
```

```python
# Define the conditions and corresponding values
conditions_celeb = [
    (df["Which celebrity's hair care philosophy resonates most with you? "] == "A minimalist tech entrepreneur"),
    (df["Which celebrity's hair care philosophy resonates most with you? "] == "A traditional cultural icon"),
    (df["Which celebrity's hair care philosophy resonates most with you? "] == "Just a chill instagram model"),
    (df["Which celebrity's hair care philosophy resonates most with you? "] == "Dermatologists")
]

choices_celeb = [
    "Tech-Savvy Innovators",
    "Cultural Heritage Influencers",
    "Laid-Back Trendsetters",
    "Science Experts"
]

# Apply np.select to create the new column based on the conditions
df['celeb_hair_care_philosophy'] = np.select(conditions_celeb, choices_celeb, default='Other')

# Display the resulting DataFrame
df[['Which celebrity\'s hair care philosophy resonates most with you? ', 'celeb_hair_care_philosophy']]
```

```python
# Create the new column based on the conditions
df['current_natural_shampoo_user'] = df['Which hair care products do you currently use?'].apply(
    lambda x: True if x == "Naturals (Life Bouy Herbal Strong, Chiltan Pure, etc)" else ("Not Specified" if x == "" else False)
)

# Display the resulting DataFrame
df[['Which hair care products do you currently use?', 'current_natural_shampoo_user']]
```

```python
# Define the conditions and corresponding values
conditions_motivation = [
    (df['What would convince you to try Herbal Essences?  '] == "Affordable Pricing"),
    (df['What would convince you to try Herbal Essences?  '] == "Botanical Certification"),
    (df['What would convince you to try Herbal Essences?  '] == "Recommendation from Trusted Source"),
    (df['What would convince you to try Herbal Essences?  '] == "Scientific Proof of Effectiveness")
]

choices_motivation = [
    "Afordability",
    "Natural Product Appeal",
    "Trust-Based Influence",
    "Evidence-Driven Motivation"
]

# Apply np.select to create the new column based on the conditions
df['motivation_to_try_herbal_essence'] = np.select(conditions_motivation, choices_motivation, default='Other')

# Display the resulting DataFrame
df[['What would convince you to try Herbal Essences?  ', 'motivation_to_try_herbal_essence']]
```

```python
# Define the conditions and corresponding values
conditions_price = [
    (df['Acceptable Price Range for Herbal Essences  (Pantene 185 ml is pkr 400 )'] == "350-450 PKR"),
    (df['Acceptable Price Range for Herbal Essences  (Pantene 185 ml is pkr 400 )'] == "450-550 PKR"),
    (df['Acceptable Price Range for Herbal Essences  (Pantene 185 ml is pkr 400 )'] == "550-650 PKR"),
    (df['Acceptable Price Range for Herbal Essences  (Pantene 185 ml is pkr 400 )'] == "650-750 PKR"),
    (df['Acceptable Price Range for Herbal Essences  (Pantene 185 ml is pkr 400 )'] == "Over 750 PKR")
]

choices_price = [
    "Budget-Conscious Consumers",
    "Mid-Range Value Seekers",
    "Quality-Oriented Buyers",
    "Premium Product Enthusiasts",
    "Luxury Brand Believers"
]

# Apply np.select to create the new column based on the conditions
df['price_sensitivity'] = np.select(conditions_price, choices_price, default='Other')

# Display the resulting DataFrame
df[['Acceptable Price Range for Herbal Essences  (Pantene 185 ml is pkr 400 )', 'price_sensitivity']]
```
