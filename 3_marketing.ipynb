```python

def segment_market(data_file):
    """Identify key market segments"""
    df = pd.read_csv(data_file)
    segments = {
        'product_messages': df['ideal_hair_care_product_message'].value_counts(),
        'influencer_groups': df['celeb_hair_care_philosophy'].value_counts(),
        'motivations': df['motivation_to_try_herbal_essence'].value_counts()
    }
    return segments

data_file = 'marketing_branding_insights.csv'
print("Market Segments:")
print(segment_market(data_file))
```

```python
import matplotlib.pyplot as plt

# Define the segments variable
segments = segment_market(data_file)

# Create bar charts for each segment
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot product messages
axes[0].bar(segments['product_messages'].index, segments['product_messages'].values)
axes[0].set_title('Product Messages')
axes[0].set_xlabel('Message')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)
for tick in axes[0].get_xticklabels():
    tick.set_ha('right')

# Plot influencer groups
axes[1].bar(segments['influencer_groups'].index, segments['influencer_groups'].values)
axes[1].set_title('Influencer Groups')
axes[1].set_xlabel('Group')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=45)
for tick in axes[1].get_xticklabels():
    tick.set_ha('right')

# Plot motivations
axes[2].bar(segments['motivations'].index, segments['motivations'].values)
axes[2].set_title('Motivations')
axes[2].set_xlabel('Motivation')
axes[2].set_ylabel('Count')
axes[2].tick_params(axis='x', rotation=45)
for tick in axes[2].get_xticklabels():
    tick.set_ha('right')

plt.tight_layout()
plt.show()
```
