```python
# Inputs
nobs1 = 72  # Total observations in Wellness and Quality Seekers group
nobs2 = 110  # Total observations in Budget-Conscious group
p1 = 34.6 / 100  # Percentage as a proportion for group 1
p2 = 22.7 / 100  # Percentage as a proportion for group 2

# Calculate the actual counts of "Yes" responses
count1 = int(nobs1 * p1)
count2 = int(nobs2 * p2)

# Perform two-proportion z-test
from statsmodels.stats.proportion import proportions_ztest

counts = [count1, count2]
nobs = [nobs1, nobs2]

stat, pval = proportions_ztest(counts, nobs)

# Conclusion based on p-value
alpha = 0.05  # Significance level

if pval < alpha:
    print("The difference in proportions is statistically significant.")
    print("Wellness and Quality Seekers are more likely to pay more than 550 compared to Budget-Conscious Group.")
else:
    print("The difference in proportions is not statistically significant.")
    print("There is no evidence to suggest Wellness and Quality Seekers are more likely to pay more than 550.")

print(f'p-value = {pval}')
```
